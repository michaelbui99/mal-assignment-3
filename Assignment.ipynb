### Checking for missing data

#### Meso

```python
from pandas.io.formats.info import DataFrameInfo

def get_missing_count_ratio_df(df: pd.DataFrame):
    row_count, col_count = df.shape
    info = DataFrameInfo(data = df)
    info_df= pd.DataFrame(
        {'Non-Null Count': info.non_null_counts, 'Dtype': info.dtypes}
    )

    # Calculating missing data per column. 
    info_df['Missing Count'] = row_count- info_df['Non-Null Count']
    info_df['Missing Ratio'] = (info_df['Missing Count'] / row_count).astype(float)
    # Sorting missing data from highest % to lowest %
    return info_df.sort_values(by=['Non-Null Count'], ascending=True)
```

### 1.b. Selecting ws/wd signals

We will prioritize having enough coverage from the signals. This can be done qualitatively by taking a look at the plots of the different signals and then spot any significant 


```python
def plot_mast_data(df: pd.DataFrame, location: str = ""):
    x_axis = df.index.values
    for col in df.columns.values:
        if (col.endswith('qc')):
            continue
        plt.figure(figsize=(15,5))
        plt.scatter(x=x_axis, y=df[col].values)
        plt.title(f'Mast - {location} {col}')
        plt.show()

def plot_meso_data(df: pd.DataFrame, location: str = "", x=None, ignore_cols = []):
    x_axis = x if x is not None else df.index.values
    ignore_cols = set(ignore_cols)
    for col in df.columns.values:
        if col in ignore_cols:
            continue
        plt.figure(figsize=(15,5))
        plt.scatter(x=x_axis, y=df[col].values)
        plt.title(f'Meso - {location} {col}')
        plt.show()
```

### 1.d. Convert the mast data from DK time to UTC time (corresponding to the time zone used in the meso data). Remember to account for summer-time in DK.
    

#### Converting mast data into UTC with Danish TZ

```python
from datetime import datetime

def account_for_daylight_savings(dt_arr):
    return list(map(lambda x: x + pd.DateOffset(hours=-1) if x < datetime(year=x.year, month=3, day=26, hour=2) and x > datetime(year=x.year, month=10, day=29, hour=3) else x + pd.DateOffset(hours=-2), dt_arr))

def localize_dt_index_to_danish_utc(df: pd.DataFrame):
    copy = df.copy();
    index_with_correct_offsets = account_for_daylight_savings(copy.index)
    index_as_strings = list(map(lambda x: str(x), index_with_correct_offsets))
    tz_aware = pd.to_datetime(pd.Series(index_as_strings), utc=True)
    copy.index = tz_aware
    copy.index = copy.index.tz_convert('Europe/Copenhagen')
    return copy
```

### 1.e. Resample the mast dataset to have the same frequency as the meso data. The meso data has one record for each hour, the mast data has one record for each 10 min.


```python
## Resampling of wind speed and wind direction 
from scipy.stats import circmean

# We use the circular mean instead of arithemtic mean to account for 
# the circular nature of the turbine direction. E.g. it could rotate from 5 deg -> 0 deg -> 355 deg
def circmean_aggregation(arr: np.array):
    return np.rad2deg(circmean(np.deg2rad(arr)))

def resample_mast_data(df: pd.DataFrame, wind_speed_column: str, wind_direction_column: str) -> pd.DataFrame:
    df_resampled = pd.DataFrame()
    df_copy = df.copy()

    # Resampling wind speed
    df_copy = df_copy.resample('1H').mean()
    df_resampled[wind_speed_column] = df_copy[wind_speed_column]

    # Resampling wind direction
    df_copy = df.copy()
    df_copy = df_copy.resample('1H').apply(circmean_aggregation)
    df_resampled[wind_direction_column] = df_copy[wind_direction_column]

    return df_resampled
```

```python
mast_borglum_signals = resample_mast_data(mast_borglum_signals, wind_direction_column='wd32', wind_speed_column='ws32')
mast_risoe_signals  = resample_mast_data(mast_risoe_signals, wind_direction_column='wd77', wind_speed_column='ws77')
```

### 1.f. Find the overlapping timestamps between the meso data and the resampled mast data. You only want to consider data in this overlapping time period in your training.

```python
def stringify_arr_elements(xs: np.array) -> list[str]:
    return list(map(lambda x: str(x), xs))

def get_overlapping_rows(meso_df: pd.DataFrame, mast_df: pd.DataFrame):
    meso_copy = meso_df.copy()
    mast_copy = mast_df.copy()

    meso_copy['t_str'] = stringify_arr_elements(meso_copy.index)
    mast_copy['t_str'] = stringify_arr_elements(mast_copy.index.tz_convert("UTC"))
    meso_signals = meso_copy.loc[meso_copy['t_str'].isin(mast_copy['t_str'])]
    mast_signals = mast_copy.loc[mast_copy['t_str'].isin(meso_copy['t_str'])]

    return meso_signals.drop('t_str', axis=1, inplace=False), mast_signals.drop('t_str',axis=1, inplace=False)
```

```python
meso_risoe_signals, mast_risoe_signals = get_overlapping_rows(meso_df=meso_risoe_signals, mast_df=mast_risoe_signals)
```

##### Plotting the wind speed means to check if linear assumption when interpolating is fair.

```python
from sklearn.linear_model import LinearRegression

x_axis = ['WSP060', 'WSP080','WSP100','WSP120','WSP140','WSP160','WSP180','WSP200','WSP220']
y_axis = data_meso_risoe[x_axis].mean()

lr = LinearRegression()
lr.fit(np.array([x for x in range(len(x_axis))]).reshape(-1, 1), y_axis.to_numpy().reshape(-1, 1))

plt.figure(figsize=(15,5))
plt.scatter(x_axis, y_axis)
plt.axline(xy1=(0, lr.intercept_[0]), slope=lr.coef_[0][0])
plt.title(f"R2: {lr.score(np.array([x for x in range(len(x_axis))]).reshape(-1,1), y_axis.to_numpy().reshape(-1,1))}")
plt.show()
```
